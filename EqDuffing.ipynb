## Equação de Duffing amortecida e forçada

Apresentada pela expressão:

$$\ddot{x}+c\dot{x}+ax + bx^3 = \lambda cos(\Omega t) + \xi$$

Onde:
- $\ddot{x}$ representa a aceleração da variável $x$ em relação ao tempo.
- $c\dot{x}$, chamado de amortecimento, onde $\dot{x}$ é a velocidade e $c$ é o coeficiente de amortecimento que determina o quão rapidamente a energia do sistema é dissipada.
- $ax + bx^3$, são os termos de rigidez não linear.
  - $ax$ representa uma parte linear da força restauradora.
  - $bx^3$ é o termo não linear. A presença de $x^3$ é o que caracteriza esta como a equação de Duffing. O sinal de $b$ determina a natureza da não linearidade:

    - Se $b>0$, o sistema é dito ter uma rigidez "endurecedora" (hardening spring).
    - Se $b<0$, o sistema é dito ter uma rigidez "amolecedora" (softening spring).

- $\lambda cos(\Omega t)$: Este é o termo de forçamento externo periódico.
  - $\lambda$ é a amplitude da força de excitação.
  - $\Omega$ é a frequência da força de excitação.

- $\xi$ é o termo que representa um ruído estocástico ou um termo de força aleatória. Em sistemas físicos reais, sempre há alguma forma de ruído que pode ser modelado desta forma.

### Aplicações
A equação de **Duffing** é fundamental no estudo de sistemas dinâmicos não lineares. Ela é usada para modelar uma ampla gama de fenômenos em diversas áreas, incluindo:

**Engenharia Mecânica:** Vibradores, osciladores, vigas sujeitas a grandes deflexões.

**Engenharia Elétrica:** Circuitos RLC não lineares, como aqueles com indutores saturáveis.

**Física:** Dinâmica de pêndulos com grandes amplitudes, ressonância não linear, caos.

**Biologia:** Modelagem de certos sistemas biológicos onde a não linearidade é importante.



### Importância

A importância da equação de Duffing reside no fato de que, apesar de sua aparente simplicidade, ela exibe uma rica variedade de comportamentos dinâmicos, como:

- **Ressonância não linear:** O pico de ressonância pode se curvar para a esquerda ou direita dependendo do tipo de não linearidade.

- **Salto (jump phenomenon):** A amplitude da resposta pode "saltar" abruptamente entre diferentes valores para pequenas variações na frequência de forçamento.

- **Oscilações harmônicas e sub-harmônicas:** O sistema pode responder em frequências que são múltiplos ou submúltiplos da frequência de forçamento.

- **Caos:** Para certas combinações de parâmetros, o sistema pode exibir comportamento caótico, o que significa que pequenas variações nas condições iniciais levam a grandes diferenças no comportamento a longo prazo.

- **Múltiplas soluções estáveis (multi-estabilidade):** Para um dado conjunto de parâmetros, o sistema pode ter mais de uma solução estável coexistindo, dependendo das condições iniciais.

## Curvas de Energia (Hamiltoniana)

$$\ddot{x} + ax + bx^3$$
$$H = V + \Pi = \frac{1}{2}\dot{x}^2 + \frac{1}{2}ax^2 + \frac{1}{4}bx^4 = E$$

Principais Características e Comportamento da equação:

Trata-se de uma forma simplificada da equação de Duffing, onde:

- $\ddot{x}$: A aceleração.
- $ax+bx^3$: Representa a força restauradora. Não há amortecimento ($c=0$) nem forçamento externo ($\lambda=0$, $xi=0$).
  - $ax$: O termo linear da força. Se a>0, ele tenta puxar o sistema de volta ao equilíbrio em x=0.
  - $bx^3$: O termo não linear. A presença de $x^3$ é o que torna a equação **"de Duffing"** e introduz uma série de comportamentos interessantes que não seriam possíveis com apenas o termo linear.



Classificação baseada no sinal de b:
O comportamento da equação de Duffing autônoma é amplamente determinado pelos sinais de a e b.

1 - Oscilador de Duffing com "mola endurecedora" (Hardening Spring): Isso ocorre quando $b>0$ (e $a>0$).
- A força restauradora cresce mais rapidamente com o deslocamento do que em um oscilador linear.

- As frequências de oscilação aumentam com a amplitude. Ou seja, quanto maior a amplitude, mais rápido o sistema oscila.

- As órbitas no espaço de fase (gráfico de $\dot{x}$ versus $x$) são fechadas e simétricas, mas suas formas são distorcidas em comparação com as elipses de um oscilador harmônico simples.

**$a=1$, $b=1$ (Mola Endurecedora - Hardening Spring)**

Neste caso, o potencial $V(x)=\frac{1}{2}x^2 + \frac{1}{4}x^4$ é sempre positivo e cresce rapidamente para grandes valores de $∣x∣$. As curvas de energia são fechadas e se parecem com elipses distorcidas que se tornam mais **"quadradas"** para energias maiores. Isso indica que as oscilações são sempre limitadas e a frequência aumenta com a amplitude, característico de uma **"mola endurecedora"**.

```python
import numpy as np
import matplotlib.pyplot as plt

# Função Hamiltoniana
def hamiltonian(x, x_dot, a, b):
    """Calcula a Hamiltoniana para a equação de Duffing."""
    return 0.5 * x_dot**2 + 0.5 * a * x**2 + 0.25 * b * x**4
```

```python
# Intervalo para x e x_dot para plotagem
# Usaremos 200 pontos para uma visualização suave das curvas
x_range = np.linspace(-3, 3, 200)
x_dot_range = np.linspace(-3, 3, 200)
```

```python
# Cria uma grade de pontos (X, Y) para calcular a Hamiltoniana
X, Y = np.meshgrid(x_range, x_dot_range)
```

```python
# Calcula o valor da Hamiltoniana Z para cada ponto (X, Y) na grade
Z = hamiltonian(X, Y, 1, 1)

# Cria uma nova figura para cada gráfico
plt.figure(figsize=(8, 6))

# Plotagem das curvas de contorno (linhas de energia constante)
# 'levels=20' para definir o número de curvas de energia a serem plotadas
# 'cmap='viridis'' para especificar o mapa de cores
contour = plt.contour(X, Y, Z, levels=20, cmap='viridis')

# Especificação de uma barra de cores para indicar os valores de energia
plt.colorbar(contour, label='Energia (H)')

# Rótulos dos eixos
plt.xlabel('x')
plt.ylabel('$\dot{x}$')

# Título do gráfico
plt.title(f'Curvas de Energia (Hamiltoniana) para a=1, b=1 (Mola Endurecedora - Hardening Spring)')

# Adiciona uma grade e fecha o gráfico
plt.grid(True)
plt.show()
```

2. Oscilador de Duffing com "mola amolecedora" (Softening Spring): Isso ocorre quando $b<0$ (e $a>0$).

- A força restauradora cresce mais lentamente com o deslocamento do que em um oscilador linear (ou até diminui para grandes deslocamentos).

- As frequências de oscilação diminuem com a amplitude. Quanto maior a amplitude, mais lento o sistema oscila.

- Para certos valores de a e b, e dependendo das condições iniciais, podem surgir poços de potencial múltiplos. Isso significa que o sistema pode ter múltiplos pontos de equilíbrio estável.

- Se $a>0$ e $b<0$, para grandes amplitudes, a força restauradora $ax+bx^3$ pode mudar de sinal, o que leva a múltiplos pontos de equilíbrio (além de $x=0$).

**$a=1$, $b=-1$ (Mola Amortecedora - Softening Spring)**

Aqui, o potencial $V(x)=\frac{1}{2}x^2 - \frac{1}{4}x^4$. Para pequenos valores de $∣x∣$, o termo $x^2$ domina e o potencial tem uma forma de poço. No entanto, para grandes valores de $∣x∣$, o termo $−x^4$ domina, fazendo com que o potencial decaia para $−\infty$. Isso significa que existem regiões de oscilação limitada (curvas fechadas ao redor da origem), mas também regiões onde as trajetórias podem escapar para o infinito (curvas abertas). As curvas fechadas mostram uma diminuição da frequência com o aumento da amplitude, característico de uma "mola amolecedora".

```python
# Calcula o valor da Hamiltoniana Z para cada ponto (X, Y) na grade
Z = hamiltonian(X, Y, 1, -1)

# Cria uma nova figura para cada gráfico
plt.figure(figsize=(8, 6))

# Plotagem das curvas de contorno (linhas de energia constante)
# 'levels=20' para definir o número de curvas de energia a serem plotadas
# 'cmap='viridis'' para especificar o mapa de cores
contour = plt.contour(X, Y, Z, levels=20, cmap='viridis')

# Especificação de uma barra de cores para indicar os valores de energia
plt.colorbar(contour, label='Energia (H)')

# Rótulos dos eixos
plt.xlabel('x')
plt.ylabel('$\dot{x}$')

# Título do gráfico
plt.title(f'Curvas de Energia (Hamiltoniana) para a=1, b=-1 (Mola Amortecedora - Softening Spring)')

# Adiciona uma grade e fecha o gráfico
plt.grid(True)
plt.show()
```

3. Outros Casos (com a=0 ou a<0):

- Se $a=0$, a equação se torna $\ddot{x} + bx^3=0$. O ponto de equilíbrio é $x=0$. Se $b>0$, ainda temos um tipo de mola endurecedora. Se $b<0$, o $x=0$ se torna um ponto de equilíbrio instável e as soluções podem divergir.
- Se $a<0$ e $b>0$, o $x=0$ é um ponto de sela (instável), e o sistema terá dois pontos de equilíbrio estável (um para $x>0$ e outro para $x<0$), além de um ponto de sela em $x=0$.

**$a=-1$ e um máximo local em x=0.**
O potencial é $V(x)=−\frac{1}{2}x^2 + \frac{1}{4}x^4$. Neste caso, o potencial tem dois mínimos locais em $x = \pm\sqrt{a/b} = \pm 1$ $\pm1$ e um máximo local em $x=0$. Isso cria um cenário de "duplo poço".

As curvas de energia revelam três tipos de comportamento:

- Oscilações em torno de um dos poços (curvas fechadas dentro de cada "lóbulo").

- Oscilações que atravessam ambos os poços (curvas maiores que envolvem os dois lóbulos).

- Uma separatriz (a curva que passa pelo ponto de sela em (0,0)), que divide os movimentos que ficam confinados a um poço dos movimentos que saltam entre os poços.

```python
# Calcula o valor da Hamiltoniana Z para cada ponto (X, Y) na grade
Z = hamiltonian(X, Y, -1, 1)

# Cria uma nova figura para cada gráfico
plt.figure(figsize=(8, 6))

# Plotagem das curvas de contorno (linhas de energia constante)
# 'levels=20' para definir o número de curvas de energia a serem plotadas
# 'cmap='viridis'' para especificar o mapa de cores
contour = plt.contour(X, Y, Z, levels=20, cmap='viridis')

# Especificação de uma barra de cores para indicar os valores de energia
plt.colorbar(contour, label='Energia (H)')

# Rótulos dos eixos
plt.xlabel('x')
plt.ylabel('$\dot{x}$')

# Título do gráfico
plt.title(f'Curvas de Energia (Hamiltoniana) para a=-1, b=1 (Potencial Duplo Poço)')

# Adiciona uma grade e fecha o gráfico
plt.grid(True)
plt.show()
```

**$a=-1$, $b=-1$ (Instável)**

O potencial é $V(x)=−\frac{1}{2}x^2 − \frac{1}{4}x^4$. Neste caso, o potencial é sempre negativo (exceto em $x=0$) e decai rapidamente para $-\infty$ à medida que $∣x∣$ aumenta. O ponto $x=0$ é um ponto de máximo local (instável), e não há poços de potencial estáveis. As curvas de energia mostram que todas as trajetórias (exceto a trivial $x=\dot{x}=0$) são ilimitadas e o sistema é instável.

```python
# Calcula o valor da Hamiltoniana Z para cada ponto (X, Y) na grade
Z = hamiltonian(X, Y, -1, -1)

# Cria uma nova figura para cada gráfico
plt.figure(figsize=(8, 6))

# Plotagem das curvas de contorno (linhas de energia constante)
# 'levels=20' para definir o número de curvas de energia a serem plotadas
# 'cmap='viridis'' para especificar o mapa de cores
contour = plt.contour(X, Y, Z, levels=20, cmap='viridis')

# Especificação de uma barra de cores para indicar os valores de energia
plt.colorbar(contour, label='Energia (H)')

# Rótulos dos eixos
plt.xlabel('x')
plt.ylabel('$\dot{x}$')

# Título do gráfico
plt.title(f'Curvas de Energia (Hamiltoniana) para a=-1, b=-1 (Instável)')

# Adiciona uma grade e fecha o gráfico
plt.grid(True)
plt.show()
```
