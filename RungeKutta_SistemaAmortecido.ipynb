# $\ddot{x} + c\dot{x} + ax + bx^3 = 0$

Podemos transformar o oscilador de Duffing com amortecimento linear c no sistema de primeira ordem para $x=t$ e $y = \dot{x}$.
$$\left\{ \begin{array}{cl}
\dot{x} = y, \\
\dot{y} = -cy - ax - bx^4
\end{array} \right.$$

# Pontos de equilíbrio (fixos)
Os pontos de equilíbrio satisfazem $y=0$ e $-ax - bx^3 = 0$, isto é
$$x(a + bx^2) = 0$$
Portanto os equilíbrios são:
$$x_0 = 0 \mbox{ e } x = \pm \sqrt{-\frac{a}{b}}\mbox{ (quando -a/b > 0),}$$
sempre com $y=0$. Logo a depender dos sinais de $a$ e $b$ teremos:
- Se $-a/b \leq 0$ ($a$ e $b$ tem o mesmo sinal ou a = 0) terá equilíbrio real (0,0).
- Se $-a/b>0$ ($a$ e $b$ tem sinais opostos) existem equilíbrios reais em (0,0) e $(\pm \sqrt{-a/b},0)$.

# Linearização e classificação local

A jacobiana do sistema é
$$\left( \begin{matrix}
0 & 1 \\
-a - 3bx^2 & -c
\end{matrix} \right)$$

## No equilíbrio trivial (0,0)
$J(0,0) = \left( \begin{matrix}
0 & 1 \\
-a & -c
\end{matrix} \right)$
Tendo por polinômio característico a expressão $\lambda^2 + c\lambda + a = 0$,
e raízes em

$$\lambda = \frac{-c \pm \sqrt{c^2 - 4a}}{2}$$


### Para c > 0 (amortecimento físico):
- a > 0: produto das raízes = a> 0 e soma = -c< 0 $\Rightarrow $ ambas as partes reais negativas $\Rightarrow $ (0,0) é assintóticamente estável (nó ou foco dependendo de $c^2 - 4a$.
- a < 0: produto = a < 0 $\Rightarrow $ autovalores de sinais opostos $\Rightarrow $ (0,0) é sela (instável).
- a = 0: caso limite com $\lambda(lambda + c)=0 \to $ um autovalor zero. Nesse caso há necessidade de análise não-linear (bifurcaçã).
Observação: Se c = 0 (sem amortecimento), o caso a > 0 dá autovalores puramente imaginários $\lambda = \pm i \sqrt{a} \to$ centro linear (conservativo).

### Nos equilíbrio não triviais $x^* = \pm \sqrt{-a/b}$

Quando existem, avalia-se a jacobiana:

$$-a - 3b(x^*)^2 = -a - 3b(-\frac{a}{b}) = -a + 3a = 2a$$,

logo,

$$ J(x^*,0) = \begin{pmatrix}
0 & 1 \\
2a & -c
\end{pmatrix}$$,

sendo seu polinômio característico

$$\lambda^2 + c\lambda - 2a = 0,$$

com raízes

$$\lambda = \frac{-c \pm \sqrt{c^2 + 8a}}{2}$$

- Note que os equilíbrios não triviais só existem quando -a/b > 0.
- Se esses equilíbrios existem e a < 0 (caso típico quando b > 0), então -2a > 0 e a soma das raízes é -c < 0 e o produto é -2a > 0. Assim ambas têm parte real negativa quando c > 0 $\to$ equilíbrios assintóticamente estáveis (nó ou foco conforme o discriminante).
- Se a > 0 e b < 0 (equivalente situação invertida), a análise é análoga à situação anterior.

Observação: Quando os ramos não triviais existes (-a/b > 0) eles costumam ser estáveis (para c>0) se o equilíbrio trivial tornou-se instável. Bifurcação do tipo pitchfork.

### Bifurcação: pitchfork (simétrico)

O sistema tem simetria ímpar $x \mapsto -x$. Á medida que variamos a (tomand b e c fixos), tem-se:

- Quando b > 0
  - Para a > 0: apenas (0,0) existe e é estável.
  - Em a = 0: ocorre uma mudança qualitativa.
  - Para a < 0: (0,0) torna-se instável (sela) e aparecem dois equilíbrios simétivos $\pm \sqrt{-a/b}$ estáveis. Isto é, a bifurcação pitchfork supercrítica - ramos estáves sugem a partir do trivial que perde estabilidade.
- Quando b < 0:
  - A criação de equilíbrios não triviais se dá para a > 0, e os ramos podem ser instáveis, refletindo um pitchform subcrítico (ramos instáveis que desaparecem quando passam por a = 0). Em aplicações físicas, esse caso pode indicar instabilidade de grande amplitude.
 
Em termos intuitivos: o termo cúbico $bx^3$ determina se o acrescimo não-linear "endurece" (b>0) ou "amolece" (b<0) a mola, o sinal de $a$ desloca o mínimo do potencial e, atravessando $a=0$, o número de mínimos do potencial pode mudar, produzindo a bifurcação.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

# ======================================================
# Definição do sistema de Duffing com amortecimento
# ======================================================
def duffing(t, state, a, b, c=0.1):
    x, y = state
    dxdt = y
    dydt = -c*y - a*x - b*x**3
    return np.array([dxdt, dydt])

# ======================================================
# Integração Runge-Kutta 4ª ordem (passo fixo)
# ======================================================
def rk4(f, t0, y0, t_end, dt, args=()):
    t_values = np.arange(t0, t_end, dt)
    y_values = np.zeros((len(t_values), len(y0)))
    y = np.array(y0, dtype=float)
    for i, t in enumerate(t_values):
        y_values[i] = y
        k1 = f(t, y, *args)
        k2 = f(t + dt/2, y + dt/2*k1, *args)
        k3 = f(t + dt/2, y + dt/2*k2, *args)
        k4 = f(t + dt, y + dt*k3, *args)
        y = y + dt*(k1 + 2*k2 + 2*k3 + k4)/6
    return t_values, y_values

# ======================================================
# Parâmetros e condições iniciais
# ======================================================
parametros = [(-1,-1), (-1,1), (1,-1), (1,1)]
iniciais = [(1,0), (0.1,0), (2,0)]
dt = 0.01
t_max = 100

# Criar pasta para salvar resultados
outdir = "duffing_rk4_outputs"
os.makedirs(outdir, exist_ok=True)

# ======================================================
# Execução: integrar e salvar resultados
# ======================================================
fig, axes = plt.subplots(len(parametros), 2, figsize=(12, 12))
summary = []

for idx, (a,b) in enumerate(parametros):
    ax_time = axes[idx,0]
    ax_phase = axes[idx,1]
    
    for x0, y0 in iniciais:
        t_vals, traj = rk4(duffing, 0, [x0,y0], t_max, dt, args=(a,b))
        x_vals, y_vals = traj[:,0], traj[:,1]

        # Salvar CSV individual
        fname = f"traj_a{a}_b{b}_x0{x0}_y0{y0}.csv"
        path = os.path.join(outdir, fname.replace(".","p"))
        pd.DataFrame({"t":t_vals,"x":x_vals,"y":y_vals}).to_csv(path, index=False)
        
        # Gráfico x(t)
        ax_time.plot(t_vals, x_vals, label=f"x0={x0}, y0={y0}")
        # Gráfico fase
        ax_phase.plot(x_vals, y_vals, label=f"x0={x0}, y0={y0}")
        
        summary.append({"a":a,"b":b,"x0":x0,"y0":y0,"file":path})
    
    ax_time.set_title(f"(a={a}, b={b}) - x(t)")
    ax_time.set_xlabel("t")
    ax_time.set_ylabel("x")
    ax_time.legend(fontsize=8)
    
    ax_phase.set_title(f"(a={a}, b={b}) - Plano de fase")
    ax_phase.set_xlabel("x")
    ax_phase.set_ylabel("y")
    ax_phase.legend(fontsize=8)

plt.tight_layout()
plt.savefig(os.path.join(outdir, "rk4_duffing_sets.png"), dpi=300)

# CSV resumo das trajetórias
pd.DataFrame(summary).to_csv(os.path.join(outdir,"summary_trajectories.csv"), index=False)

print("Simulações concluídas. Resultados salvos na pasta:", outdir)
```
